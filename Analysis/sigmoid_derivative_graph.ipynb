$$f(x) = x \cdot (1 - x)$$

The sigmoid derivative, $$\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))$$, has an elegant mathematical property. It connects the function back to itselfâ€”its slope depends on its value at a given point. This simplicity is one of the reasons sigmoid derivatives are popular in neural networks for activation gradients during backpropagation.

```python
def sigmoid_derivative(x):
    return x * (1 - x)
```

## Define the function

```python
def f(x):
    return x * (1 - x)
```

## Generate x values
Values between 0 and 1

```python
x = np.linspace(0, 1, 100) 
x
```

## Compute y values

```python
y = f(x)
```

## Plot the function

```python
plt.plot(x, y, label='y = x * (1 - x)')
plt.title('Plot of y = x * (1 - x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
